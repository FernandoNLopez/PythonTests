```python
#crear una variable root para los links
root = "https://www.football-data.co.uk/mmz4281/"

#crear lista de ligas
leagues = ['E0', 'E1']

#crear el loop en la lista de ligas, leer los csv y adjuntarlos en una variable con el método append.
#Hacer un loop para extraer las temporadas e insertarlas en una nueva columna

frames = []

for league in leagues:
   for season in range(15, 25):
      df = pd.read_csv(root + str(season) + str(season + 1) + '/' + league + '.csv', encoding='unicode_escape')
      df.insert(1, 'season', season)

      frames.append(df)
```

```python
#Organizar data en un Diccionario 

dict_countries = {
    'Spanish La Liga' : 'SP1', 'Spanish Segunda Division' : 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League' : 'E0', 'English Championship' : 'E1'
}
```

```python
#crear un loop sobre las llaves del diccionario

dict_historico = {}


for league in dict_countries:
    frames = [] # un dataframe por liga

    for season in range(20, 25):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league]+ ".csv")
        df.insert(1, 'season', season)
        frames.append(df)

    df_frames = pd.concat(frames) # almacenar todas las temporadas de la liga en un solo dataframe
    dict_historico[league] = df_frames # asignar dataframe a un nuevo diccionario
```
