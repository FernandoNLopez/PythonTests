```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='unicode_escape')
```

                                    
                                    Encontrar Duplicados

```python
#en una columna o serie
df_laptops[df_laptops.duplicated('laptop_ID')]
```

```python
#duplicados en 2 o mas columnas
duplicated = df_laptops.duplicated(['Product', 'TypeName', 'Inches'])
```

```python
df_laptops[duplicated].sort_values(['Product', 'TypeName'])
```

```python
#Ordenar asc por company y price (barato primeros)
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
#Cantidad de duplicados
df_laptops.value_counts('Company')
```

```python
#valores de duplicados en la columna Company (por defecto --> keep='first')
duplicated_first = df_laptops.duplicated('Company')
```

```python
#Mostrar dataframe con valores duplicados
df_laptops[duplicated_first]
```
