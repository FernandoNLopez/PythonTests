```python
df_cars.mean(numeric_only=True)
```

```python
df_cars.agg({
    'Sales_in_thousands':'mean',
    'Price_in_thousands':'mean',
    'Engine_size':'mean',
    'Horsepower':'mean',
    'Fuel_capacity':'mean'
})
```

```python
df_cars.groupby('Manufacturer').agg({
    'Sales_in_thousands': ['sum', 'mean'],
    'Price_in_thousands': ['sum', 'mean']
})
```
