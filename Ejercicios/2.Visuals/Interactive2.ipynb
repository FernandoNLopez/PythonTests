```python
import pandas as pd
import cufflinks as cf
import plotly.io as pio

# Configurar modo offline y renderer compatible con VS Code
cf.go_offline()
pio.renderers.default = 'notebook_connected'
```

```python
# Display themes
cf.getThemes()
```

```python
#declare dataframe
df_population = pd.read_csv('population_total.csv')
df_population
```

```python
#take out null values
df_population.dropna(inplace=True)

#reshape the dataframe
df_pop_pivot = df_population.pivot(index='year', columns='country', values='population')
```

```python
#select clause to filter countries
df_pop_pivot = df_pop_pivot[['China', 'United States','Italy', 'Argentina', 'Spain']]
df_pop_pivot
```

```python
df_pop_pivot.index = df_pop_pivot.index.astype(int)  # Asegura que el índice sea int
df_pop_pivot = df_pop_pivot.dropna(how='any')        # Elimina filas con valores nulos
```

```python
print(df_pop_pivot.head())
print(df_pop_pivot.dtypes)
```

```python
df_pop_pivot.iplot(kind='line', title='Population Evolution',
                   xTitle='year', yTitle='Population')
```
